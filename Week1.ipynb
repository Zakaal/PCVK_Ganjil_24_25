# LANGKAH (STEP 1):


```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

```

# LANGKAH (STEP 2):

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://images.ctfassets.net/cnu0m8re1exe/4NoyJgkCbQ7RYUMYjNk9v5/95b5d2f4e2024223cfb12cb43016efbe/shutterstock_299796383.jpg?fm=jpg&fl=progressive&w=660&h=433&fit=fill"]

# Baca dan tampilkan image
# Loop pada tiap url image, beberapa image dapat disimpan dalam list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

# LANGKAH (STEP 3):

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

# LANGKAH (STEP) 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis vertikal putih pada gambar
for x in range(lebar):
    if int(x) == lebar // 2:
        image_3[:, x] = [255, 255, 255]
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan gambar
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Mendapatkan dimensi gambar
tinggi, lebar, _ = image.shape

# Membuat garis diagonal putih pada gambar
for x in range(lebar):
    for y in range(tinggi):
        if x == y:
            image_3[y, x] = [255, 255, 255]

# Menggabungkan gambar
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan gambar
cv2_imshow(final_frame)
```
