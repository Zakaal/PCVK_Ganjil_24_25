# PRAKTIKUM LANGKAH (STEP 1):


```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

```

# PRAKTIKUM LANGKAH (STEP 2):

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://images.ctfassets.net/cnu0m8re1exe/4NoyJgkCbQ7RYUMYjNk9v5/95b5d2f4e2024223cfb12cb43016efbe/shutterstock_299796383.jpg?fm=jpg&fl=progressive&w=660&h=433&fit=fill"]

# Baca dan tampilkan image
# Loop pada tiap url image, beberapa image dapat disimpan dalam list
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

# PRAKTIKUM LANGKAH (STEP 3):

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)
```

# PRAKTIKUM LANGKAH (STEP) 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal ditengah image
image_3[108,:] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

# TUGAS

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menggunakan Perulangan
for x in range (tinggi):
  image_3[x, int((lebar)/2)] = [255,255,255]

# Tanpa Perulangan
image_4 = cv.line(image_4, (int(lebar/2), 0), (int(lebar/2), tinggi), (255, 255, 255), 1)

final_frame = cv.hconcat((image_2, image_3, image_4))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menggunakan Perulangan
diagonal_length = int(plt.sqrt(tinggi**2 + lebar**2))

for i in range(diagonal_length):
    x = int(i * lebar / diagonal_length)
    y = int(i * tinggi / diagonal_length)
    if x < lebar and y < tinggi:
        image_3[y, x] = [255, 255, 255]
for i in range(diagonal_length):
    x = int(i * lebar / diagonal_length)
    y = int((diagonal_length - i) * tinggi / diagonal_length)
    if x < lebar and y < tinggi:
        image_3[y, x] = [255, 255, 255]

# Tanpa Perulangan
image_4 = cv.line(image_2, (0, 0), (lebar, tinggi), (255, 255, 255), 1)
image_4 = cv.line(image_2, (lebar, 0), (0, tinggi), (255, 255, 255), 1)

final_frame = cv.hconcat((image_2, image_3, image_4))
cv2_imshow(final_frame)
```

# TUGAS KELOMPOK

```python
ktp_link = "https://umsu.ac.id/artikel/wp-content/uploads/2023/11/cara-mudah-cek-ktp-asli-atau-palsu.jpeg"
image_ktp = io.imread(ktp_link)
image_ktp = cv.resize(image_ktp, (0,0), fx=0.5, fy=0.5)
image_2 = cv.cvtColor(image_ktp, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image_ktp, cv.COLOR_BGR2RGB)

# Menutupi Nama Provinsi
for x in range (26, 36):
  for y in range (172, 240):
    image_3[x,y] = [0,255,255]

# Menutupi Nama Kota
for x in range (39, 49):
  for y in range (135, 225):
    image_3[x,y] = [255,0,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
